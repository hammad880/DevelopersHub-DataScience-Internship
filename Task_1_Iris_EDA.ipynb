# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to understand dataset structure, perform basic analysis, and visualize relationships between variables using Python libraries such as pandas, matplotlib, and seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
print("Shape:", iris.shape)
print("\nColumns:", iris.columns)
iris.info()
iris.describe()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
iris.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=iris)
plt.title("Boxplot of Iris Features")
plt.show()
```

## Conclusion

From the visualizations, we observed that different species of iris flowers have distinguishable petal and sepal measurements. Petal length and width provide strong separation between species, making them useful features for classification tasks.
